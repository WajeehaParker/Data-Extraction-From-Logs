```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Read the log file and convert it into a dataframe
logs = pd.read_csv('./API Logs.log', delimiter=' ', header=0)

# Display the dataframe
logs.head(20)
```

```python
logs = logs[['date', 'time', 'cs-method', 'cs-uri-stem', 's-port', 'sc-status', 'sc-bytes', 'cs-bytes', 'time-taken']]
logs.head()
```

```python
logs.columns
```

```python
filtered_df = logs[logs['sc-status'] == 500]
filtered_df.head(20)
```

```python
filtered_df = logs[logs['sc-status'] >= 400]
filtered_df.head(200)
```

```python
filtered_df = logs[logs['time-taken'] > 1000]
filtered_df.head(20)
```

```python
df = logs[['date', 'time', 'cs-method', 'cs-uri-stem', 's-port', 'sc-status', 'sc-bytes', 'cs-bytes', 'time-taken']]
df['hour'] = pd.to_datetime(df['time']).dt.hour
df.drop('time', axis=1, inplace=True)
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df.drop('date', axis=1, inplace=True)
```

```python
df.head()
```

```python
df_encoded = pd.get_dummies(df)
df_encoded.head()
```

```python
df_encoded.shape
```

```python
from sklearn.cluster import KMeans
```

```python
sse = [] # within cluster distance
for k in range(2, 15):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(df_encoded)
    sse.append(kmeans.inertia_)
```

```python
fig = plt.figure(figsize=(17, 5))
plt.plot(range(2, 15), sse)
plt.grid(True)
plt.title('Elbow curve')
```

```python
kmeans = KMeans(n_clusters=5)

# Fit the data to the KMeans model
kmeans.fit(df_encoded)

# Get the cluster labels for each data point
labels = kmeans.labels_
```

```python
plt.scatter(df_encoded['hour'], df_encoded['time-taken'], c=labels)
plt.xlabel('hour')
plt.ylabel('time taken')
plt.title('KMeans Clusters')
plt.show()
```
